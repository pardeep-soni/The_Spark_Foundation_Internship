#### Prediction using Supervised ML
Predicting the percentage of an student based on the no. of study hours.
This is a simple linear regression as it involves just 2 variables.


```python
# Importing neccessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing Dataset through URL

```python
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
```

#### reviewing dataset

```python
df.head(5)
```

```python
df.plot( subplots=True);
```

### seems likes the data is highly correlated

```python
# Plotting the distribution of scores
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

#### Data is positively correlated

### Preprocessing the data

### Standardising the data to make the line best fit because Linear Regression is very sensitive towards outliers

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc  = StandardScaler()
```

```python
df_sc = sc.fit_transform(df)
```

```python
df_sc = pd.DataFrame(df_sc)
```

### spliting the dependednt and independent varriable

```python
X_train = df_sc.iloc[:, :-1].values  
y_train = df_sc.iloc[:, 1].values  
```

```python
# Import train_test_split from sklearn's model_selection module
from sklearn.model_selection import train_test_split

# Use train_test_split to split X & y into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Preparing a machine learning model

```python
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
```

```python
ln.fit(X_train,y_train)
```

```python
y_pred = ln.predict(X_test)
```

```python
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# You can also test with your own data
hours = [[9.25]]
own_pred = ln.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
